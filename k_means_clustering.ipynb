```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df1 = pd.read_csv(r'C:\Users\Dell\Desktop\New folder (3)\neede exels\income_data.csv')
```

```python
df1
```

```python
plt.scatter(df1['Age'], df1['Income($)'])
```

```python
from sklearn.cluster import KMeans

km = KMeans(n_clusters=3)
```

```python
km
```

```python
y_predicted = km.fit_predict(df1[['Income($)']])
```

```python
df1['Cluster'] = y_predicted
df1
```

```python
import matplotlib.pyplot as plt

df2 = df1[df1.Cluster == 0]
df3 = df1[df1.Cluster == 1]
df4 = df1[df1.Cluster == 2]

plt.scatter(df2.Age, df2['Income($)'], color='pink', label='Cluster 0')
plt.scatter(df3.Age, df3['Income($)'], color='blue', label='Cluster 1')
plt.scatter(df4.Age, df4['Income($)'], color='yellow', label='Cluster 2')

plt.xlabel('Age')
plt.ylabel('Income($)')
plt.legend()  # Now the legend will correctly display
plt.show()
```

## MinMax Scaler

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
scaler.fit(df1[['Income($)']])
df1[['Income($)']] = scaler.fit_transform(df1[['Income($)']])
df1
```

```python
scaler.fit(df1[['Age']])
df1[['Age']] = scaler.fit_transform(df1[['Age']])
df1
```

```python
km = KMeans(n_clusters=3)
y_predicted = km.fit_predict(df1[['Age','Income($)']])
y_predicted
```

```python
df1['Cluster'] = y_predicted
df1
```

```python
df2 = df1[df1.Cluster == 0]
df3 = df1[df1.Cluster == 1]
df4 = df1[df1.Cluster == 2]

plt.scatter(df2.Age, df2['Income($)'], color='pink', label='Cluster 0')
plt.scatter(df3.Age, df3['Income($)'], color='blue', label='Cluster 1')
plt.scatter(df4.Age, df4['Income($)'], color='yellow', label='Cluster 2')

plt.xlabel('Age')
plt.ylabel('Income($)')
plt.legend()  # Now the legend will correctly display
plt.show()
```

```python
km.cluster_centers_
```

```python
df2 = df1[df1.Cluster == 0]
df3 = df1[df1.Cluster == 1]
df4 = df1[df1.Cluster == 2]

plt.scatter(df2.Age, df2['Income($)'], color='pink', label='Cluster 0')
plt.scatter(df3.Age, df3['Income($)'], color='blue', label='Cluster 1')
plt.scatter(df4.Age, df4['Income($)'], color='yellow', label='Cluster 2')

plt.scatter(km.cluster_centers_[:, 0], km.cluster_centers_[:, 1], color='black', marker='x',label='centroids')

plt.xlabel('Age')
plt.ylabel('Income($)')
plt.legend()  # Now the legend will correctly display
plt.show()
```

## Elbow PLot

```python
km_rng = range(1, 10)
sse = []
for k in km_rng:
    km = KMeans(n_clusters=k)
    km.fit(df1[['Age','Income($)']])
    sse.append(km.inertia_)
```

```python
sse
```

```python
plt.xlabel('K')
plt.ylabel('SSE')
plt.plot(km_rng, sse)
```

As Elbow is at 3,therefore 3 clusters
